```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = cv2.imread('raw_image.jpg',3)
```

```python
plt.imshow(image)
plt.show()
```

```python
y=1836
x=2000
h=240
w=600
crop = image[y:y+h, x:x+w]
plt.imshow(crop)
plt.show()
```

```python
gray = cv2.cvtColor(crop,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
plt.show()
```

```python
t , binary = cv2.threshold(gray,160,255,cv2.THRESH_BINARY)
plt.imshow(binary)
plt.show()
```

```python
cv2.imwrite('crop.jpg',binary)
```

```python
print(pytesseract.image_to_string(Image.open('crop.jpg')))
```
